```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats

stations = []
measurements = []


lenght_cf = 0
lenght_df = 0

for i in range(120):
    
    if i < 10:
        filename = "data/" + "00"+ str(i) + '/'
    elif  i < 100:
        filename = "data/" + "0"+ str(i) + '/'
    else:
        filename = "data/" + str(i) + '/'
        
    cf = pd.read_csv(filename+ "measurements.csv",sep ='\t')
    measurements.append(cf)
    df = pd.read_csv(filename+ "stations.csv",sep ='\t')
    stations.append(df)
   
    #print(cf.shape)
    #print(df.shape)
    #lenght_cf = lenght_cf + len(cf)
    #lenght_df = lenght_df + len(df)
    

merged_cf = pd.concat(measurements)
merged_df = pd.concat(stations)

print("Measurements: ",merged_cf.shape)

merged_cf.info()
    
print("\nStations: ",merged_df.shape)

merged_df.info()


```
